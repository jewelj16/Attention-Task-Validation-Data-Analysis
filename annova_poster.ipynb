```python
import matplotlib.pyplot as plt

# Increased figure height to make room for the flowchart at the top
fig, ax = plt.subplots(figsize=(12, 8.5), facecolor='white')
ax.axis('off')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Title
plt.text(0.55, 0.98, "Participant Flow & 2×2 Mixed Factorial Design", ha='center', va='center', 
         fontsize=18, fontweight='bold', color='#333333')

# Define new Y coordinates to push the matrix down and make room for the flow chart
col1_x = 0.40
col2_x = 0.75
row1_y = 0.50
row2_y = 0.25
col_label_y = 0.70
row_label_x = 0.20
flow_y = 0.88

# Style dictionaries
label_props = dict(facecolor='#f0f0f0', edgecolor='#aaaaaa', boxstyle='round,pad=0.4')
cell_props = dict(boxstyle='round,pad=0.7', edgecolor='#666666', linewidth=1.5)
# New style for flowchart boxes (a nice soft blue to distinguish them from the test conditions)
flow_props = dict(facecolor='#e0eaf5', edgecolor='#4a76a8', boxstyle='round,pad=0.6', linewidth=1.5)


# --- FLOWCHART ---

# 1. Recruitment Box
plt.text(0.20, flow_y, "1. Recruitment\n(N = 37)", 
         ha='center', va='center', fontsize=12, fontweight='bold', color='#1d3c63', bbox=flow_props)

# 2. Random Assignment Box
plt.text(0.575, flow_y, "2. Random Assignment\n(Counterbalanced Order)", 
         ha='center', va='center', fontsize=12, fontweight='bold', color='#1d3c63', bbox=flow_props)

# Arrow 1: Recruitment -> Assignment
ax.annotate('', xy=(0.40, flow_y), xytext=(0.28, flow_y),
            arrowprops=dict(arrowstyle="->", color='#4a76a8', lw=2))

# Arrow 2: Assignment -> Group A (curved line pointing down-left)
ax.annotate('', xy=(col1_x, col_label_y + 0.06), xytext=(0.53, flow_y - 0.04),
            arrowprops=dict(arrowstyle="->", color='#4a76a8', lw=2, connectionstyle="arc3,rad=0.1"))

# Arrow 3: Assignment -> Group B (curved line pointing down-right)
ax.annotate('', xy=(col2_x, col_label_y + 0.06), xytext=(0.62, flow_y - 0.04),
            arrowprops=dict(arrowstyle="->", color='#4a76a8', lw=2, connectionstyle="arc3,rad=-0.1"))


# --- HEADERS ---

plt.text(col1_x, col_label_y, "Group A\n(Single Target, n=21)", 
         ha='center', va='center', fontsize=12, fontweight='bold', color='black', bbox=label_props)

plt.text(col2_x, col_label_y, "Group B\n(Multiple Target, n=16)", 
         ha='center', va='center', fontsize=12, fontweight='bold', color='black', bbox=label_props)

plt.text(0.575, col_label_y + 0.08, "Between-Subjects: Target Load", 
         ha='center', va='center', fontsize=10, fontstyle='italic', color='#555555')

# Rows (Within-Subjects Labels)
plt.text(row_label_x, row1_y, "Modality 1\nLab / Computer", 
         ha='right', va='center', fontsize=12, fontweight='bold', color='black', bbox=label_props)

plt.text(row_label_x, row2_y, "Modality 2\nGame / Mobile", 
         ha='right', va='center', fontsize=12, fontweight='bold', color='black', bbox=label_props)

plt.text(0.05, 0.375, "Within-Subjects:\nModality", 
         ha='center', va='center', fontsize=10, fontstyle='italic', color='#555555', rotation=90)


# --- INNER CELLS ---

# Lab Row (Purple)
plt.text(col1_x, row1_y, "Condition 1\nSingle / Lab", 
         ha='center', va='center', fontsize=14, color='white', fontweight='bold',
         bbox=dict(facecolor='#9467bd', **cell_props))

plt.text(col2_x, row1_y, "Condition 2\nMultiple / Lab", 
         ha='center', va='center', fontsize=14, color='white', fontweight='bold',
         bbox=dict(facecolor='#9467bd', **cell_props))

# Game Row (Green)
plt.text(col1_x, row2_y, "Condition 3\nSingle / Game", 
         ha='center', va='center', fontsize=14, color='white', fontweight='bold',
         bbox=dict(facecolor='#2ca02c', **cell_props))

plt.text(col2_x, row2_y, "Condition 4\nMultiple / Game", 
         ha='center', va='center', fontsize=14, color='white', fontweight='bold',
         bbox=dict(facecolor='#2ca02c', **cell_props))

# Save the figure
plt.savefig('study_design_combined.png', dpi=300, bbox_inches='tight')
plt.show()
```
